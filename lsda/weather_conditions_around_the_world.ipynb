```python
# Importing libraries
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
# Display all df w/o truncation
pd.set_option('display.max_columns', None)
```

```python
# Setting up dataframe
df = pd.read_csv("data/GlobalWeatherRepository.csv")
```

```python
weather_conditions = px.histogram(df, x='condition_text', title='Weather Conditions around the World', color='condition_text')
weather_conditions.update_xaxes(title="Weather Condition")
weather_conditions.update_yaxes(title="Count")
weather_conditions.update_layout(legend_title_text="Weather Condition")

weather_conditions.update_traces(texttemplate="Count %{y:2f}", textposition="outside") # Adds count to bars, refer here: https://plotly.com/python/reference/bar/#bar-texttemplate

weather_conditions
```
